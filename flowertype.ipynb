```
from sklearn.model_selection import StratifiedKFold
```

```
X_train=X
y_train=y
X_test=test[X.columns]
```

```
skf = StratifiedKFold(10)
accuracy = []
preds=[]
i=1
skf.get_n_splits(X,y)
for train_index, test_index in skf.split(X,y):
  print(f'fold {i} : ', end='')
  X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
  y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]
 
  cbc.fit(X1_train, y1_train,eval_set=(X1_test,y1_test),verbose=False)
  labels = cbc.predict_proba(X1_test)
  labels_final = cbc.predict_proba(X_test)
  acc_score = log_loss(y1_test, labels,labels=list(range(0,8)))
  print(f'{acc_score}--->')
  accuracy.append(acc_score)
  if i==0:
    break
  else:
    i+=1
  
  preds.append(labels_final)
print(accuracy)
```
